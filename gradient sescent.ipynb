```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.array([1.00,2.00,3.00])
y=np.array([1.00,2.50,3.50])
plt.plot(x,y,label='original data',color='r')
plt.legend()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
print(x_mean,y_mean)
```

```python
#random value

m1=0.5
c=0
best_fit_1=np.array([])
for i in range(3):
    k=m1*x[i]+c
    best_fit_1=np.append(best_fit_1,np.array([k]),axis=0)
print(best_fit_1)   #first y prediction
```

```python
#second sample
m2=1.0
c=0
best_fit_2=np.array([])
for i in range(3):
    k=m2*x[i]+c
    best_fit_2=np.append(best_fit_2,np.array([k]),axis=0)
print(best_fit_2)   #first y prediction
```

```python
#third sample
m3=1.5
c=0
best_fit_3=np.array([])
for i in range(3):
    k=m3*x[i]+c
    best_fit_3=np.append(best_fit_3,np.array([k]),axis=0)
print(best_fit_3)   #first y prediction
```

```python
plt.plot(x,y,label='original data',color='r')
plt.plot(x,best_fit_1,label='best_fit_1',color='g')
plt.plot(x,best_fit_2,label='best_fit_2',color='m')
plt.plot(x,best_fit_3,label='best_fit_3',color='b')
plt.legend()
```

```python
#understanding the cost function for best_fit_1
#squared error cost function
#(1/2m)*signal(i=1 to n )(htheta(x(i))-(y(i))**2)
b1=0
for i in range(3):
    b1=b1+((best_fit_1[i]-y[i])**2)
c1=(1/6)*b1
c1
```

```python
#understanding the cost function for best_fit_2
#squared error cost function
#(1/2m)*signal(i=1 to n )(htheta(x(i))-(y(i))**2)
b2=0
for i in range(3):
    b2=b2+((best_fit_2[i]-y[i])**2)
c2=(1/6)*b2
c2
```

```python
#understanding the cost function for best_fit_3
#squared error cost function
#(1/2m)*signal(i=1 to n )(htheta(x(i))-(y(i))**2)
b3=0
for i in range(3):
    b3=b3+((best_fit_3[i]-y[i])**2)
c3=(1/6)*b3
c3
```

A lowest cost is desirable,best_fit_2 has the low cost so it is a good fit for our data
